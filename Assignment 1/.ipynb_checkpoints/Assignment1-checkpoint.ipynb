```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
from scipy.stats import t

# Current version of ChatGPT 3.5 (Feb 2024) has been used in creating the code for this assignment.  

# Set colors for consistency
color_palette = ["darkslateblue", "crimson"]
```

```python
# Exercise 1 
# Read the data from the file smoking.csv, divide into smokers and non-smokers. Compute average FEV1 for both groups. 

# read data
names = ["age", "FEV1", "height", "gender", "status", "weight"] # adding header information according to given information
df = pd.read_csv('smoking.csv', encoding='iso8859-1', sep='\t', names=names)

# split into smoker and non-smokers (status = 1 indicates smoker)
df_smokers = df[df["status"] == 1]
df_non_smokers = df[df["status"] == 0]

# find and print mean and variance of FEV1 for both groups as well as minimum age
print("Smokers FEV1: ")
print(df_smokers.loc[:, "FEV1"].mean())
print("Smoker minimum age")
print(df_smokers.loc[:, "age"].min())
print("Smoker FEV1 variance")
print(df_smokers.loc[:, "FEV1"].var(),"\n")

print("Non-smokers FEV1: ")
print(df_non_smokers.loc[:, "FEV1"].mean())
print("Non smoker minimum age")
print(df_non_smokers.loc[:, "age"].min())
print("Non smoker FEV1 variance")
print(df_non_smokers.loc[:, "FEV1"].var())
```

```python
# Exercise 2 
# Make Box plots
df['status'] = df['status'].replace({0: 'Non-smoker', 1: 'Smoker'}) # change values in dataframe to automatically get proper labels
sns.set(rc={'figure.figsize':(10,4)}, style="ticks") 
sns.set_style("whitegrid")
ax = sns.boxplot(x = "FEV1", y = "status", data=df, orient="h", hue="status", palette=color_palette)

plt.legend(loc='lower left')
plt.savefig('A1_Boxplot.png')
plt.show()


# Exercise 3
# Perform two-sided t-test
# Null hypothesis: The two groups have the same mean

# Welsch: Unequal variance (minimum age is quite different -> var: 0.56 vs. 0.72) and unequal sample sizes (n=65 vs n=589)

def t_test(a, b, alpha):
    '''Performs a Welsch t-test of the two distributions a and b using alpha as the the cutoff for significance'''
    # calculate for both distributions: mean, corrected standard deviations, number of samples, standard error
    a_x, b_x = a.mean(), b.mean() # mean
    a_n, b_n = len(a), len(b) # number of samples
    a_std, b_std = a.std(), b.std() # corrected standard deviation
    a_s, b_s = (a_std/np.sqrt(a_n)), (b_std/np.sqrt(b_n)) # standard error

    # Calculate number of degrees of freedom
    dof = ((a_std**2/a_n)+(b_std**2/b_n))**2/(a_std**4/(a_n**2 * (a_n - 1)) + b_std**4/(b_n**2 * (b_n - 1)))

    # Calculate t-stat
    t_stat = (a_x - b_x)/(np.sqrt(a_s**2 + b_s**2))

    # Calculate p-value using scipy cdf
    p_value = 2 * (1 - t.cdf(abs(t_stat), dof))

    # either accept for reject null hypothesis
    print("Results from Welsch's t-test are as following: ")
    print("dof is:", dof)
    print("t-stat is:", t_stat)
    print("p-value is:", p_value)
    if p_value < alpha:
        print("The null hypothesis can thus be rejected")
    else:
        print("Therefore the null hypothesis cannot be rejected")

alpha = 0.05 # significance level
t_test(df_non_smokers.loc[:, "FEV1"], df_smokers.loc[:, "FEV1"], alpha)

# using 'easy' method to test
# print("T-test: ")
# print(scipy.stats.ttest_ind(df_non_smokers.loc[:, "FEV1"], df_smokers.loc[:, "FEV1"], equal_var=False))
# testResult(statistic=-7.149608129503808, pvalue=3.073812744876234e-10, df=83.27311025102057) (df: degrees of freedom)
```

```python
# Exercise 4
# correlation between age and FEV1 with scatterplot
sns.set_style("whitegrid")
sns.scatterplot(x="age", y="FEV1", hue="status", data=df, palette=color_palette, s=45, alpha=0.6)
sns.set(rc={'figure.figsize':(10, 8)})

plt.savefig('A1_Scatterplot.png')
plt.show()

# correlation 
corr = df['age'].corr(df['FEV1'])
print("Pearson's correlation between age and FEV1 is:", corr) # Pearson
corr1 = df['age'].corr(df['FEV1'], method="spearman")
print("Spearman's correlation between age and FEV1 is:", corr1) # Spearman
```
